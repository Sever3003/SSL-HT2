```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import LambdaLR
from torch.utils.data import DataLoader
from torchvision.models import resnet18
from medmnist import PneumoniaMNIST
from functions import *
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score, recall_score, precision_score
import torchvision.transforms as transforms
from torchvision.transforms import InterpolationMode

# --- Гиперпараметры ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
num_classes = 2
projection_dim = 128
batch_size = 512
learning_rate = 0.001
n_epochs = 20

# --- Загружаем данные ---
dataset = PneumoniaMNIST(split='train', download=True, as_rgb=False)
train_transform = transforms.Compose([
    transforms.RandomResizedCrop(
        size=(224, 224),
        scale=(0.75, 1.0),
        ratio=(0.75, 1.0),
        interpolation=InterpolationMode.BICUBIC,
        antialias=True
    ),
    transforms.RandomApply(
        [transforms.ColorJitter(
            brightness=(0.9, 1.1),
            contrast=(0.9, 1.1),
            saturation=(0.9, 1.1),
            hue=(-0.1, 0.1)
        )],
        p=0.7 
    ),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(
        degrees=15, 
        interpolation=InterpolationMode.BILINEAR,
        fill=0
    ),
    transforms.ToTensor(),
])

test_transform = transforms.Compose([
    transforms.ToTensor(),
])


train_dataset = GenericPairDataset(dataset, transform=train_transform)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# --- Модель ---
base_encoder = resnet18
simclr_model = SimCLR(base_encoder, projection_dim).to(device)
linear_probe = LinModel(simclr_model.enc, simclr_model.feature_dim, n_classes=num_classes).to(device)

# --- Оптимизаторы ---

simclr_optimizer = torch.optim.SGD(
    simclr_model.parameters(),
    lr=0.2, 
    momentum=0.9,
    weight_decay=0.,
    nesterov=True
)

probe_optimizer = torch.optim.SGD(
    linear_probe.lin.parameters(),
    lr=0.2,
    momentum=0.9,
    weight_decay=0.,
    nesterov=True
)

simclr_scheduler = LambdaLR(
    simclr_optimizer,
    lr_lambda=lambda step: get_lr(
        step,
        n_epochs * len(train_loader),
        learning_rate,
        1e-4
    )
)

probe_scheduler = LambdaLR(
    probe_optimizer,
    lr_lambda=lambda step: get_lr(
        step,
        n_epochs * len(train_loader),
        learning_rate,
        1e-4
    )
)

def train_model(simclr_model, linear_probe, train_loader, test_loader, device, epochs):
    simclr_model.train()
    linear_probe.train()

    train_losses, test_losses = [], []
    train_f1_scores, test_f1_scores = [], []
    train_recalls, test_recalls = [], []
    train_precisions, test_precisions = [], []

    for epoch in range(epochs):
        epoch_ssl_loss = 0
        epoch_cls_loss = 0
        train_preds, train_targets = [], []

        for x, y in tqdm(train_loader):
            x, y = x.to(device), y.to(device)
            x1, x2 = x[:, 0, :, :], x[:, 1, :, :]

            # --- SimCLR Loss ---
            _, z1 = simclr_model(x1)
            _, z2 = simclr_model(x2)
            ssl_loss = nt_xent(torch.cat([z1, z2], dim=0))

            simclr_optimizer.zero_grad()
            ssl_loss.backward()
            simclr_optimizer.step()
            simclr_scheduler.step()

            # --- Classification Loss ---
            with torch.no_grad():
                h, _ = simclr_model(x1)
                h = h.detach()

            logits = linear_probe.lin(h)
            class_weights = torch.tensor([0.5, 0.5]).to(device) 
            cls_loss = nn.CrossEntropyLoss(weight=class_weights)(logits, y)
            # cls_loss = nn.CrossEntropyLoss()(logits, y)

            probe_optimizer.zero_grad()
            cls_loss.backward()
            probe_optimizer.step()
            probe_scheduler.step()

            epoch_ssl_loss += ssl_loss.item()
            epoch_cls_loss += cls_loss.item()

            train_preds.extend(torch.argmax(logits, dim=1).cpu().numpy())
            train_targets.extend(y.cpu().numpy())

        avg_ssl_loss = epoch_ssl_loss / len(train_loader)
        avg_cls_loss = epoch_cls_loss / len(train_loader)
        train_losses.append(avg_cls_loss)
        train_f1_scores.append(f1_score(train_targets, train_preds, average="binary"))
        train_recalls.append(recall_score(train_targets, train_preds, average="binary"))
        train_precisions.append(precision_score(train_targets, train_preds, average="binary"))

        # --- Оценка ---
        linear_probe.eval()
        test_preds, test_targets = [], []
        total_test_loss = 0
        with torch.no_grad():
            for x, y in tqdm(test_loader, desc=f"Epoch {epoch+1}/{epochs} - Testing"):
                x, y = x.to(device), y.to(device).squeeze().long()
                logits = linear_probe(x)
                loss = nn.CrossEntropyLoss()(logits, y)
                total_test_loss += loss.item() * x.size(0)
                test_preds.extend(logits.argmax(dim=1).cpu().numpy())
                test_targets.extend(y.cpu().numpy())

        test_losses.append(total_test_loss / len(test_loader.dataset))
        test_f1_scores.append(f1_score(test_targets, test_preds, average="binary"))
        test_recalls.append(recall_score(test_targets, test_preds, average="binary"))
        test_precisions.append(precision_score(test_targets, test_preds, average="binary"))

        print(f"Epoch {epoch + 1}/{epochs}: Train F1={train_f1_scores[-1]:.4f}, Test F1={test_f1_scores[-1]:.4f}, Train Recall={train_recalls[-1]:.4f}, Test Recall={test_recalls[-1]:.4f}")

    return train_losses, test_losses, train_f1_scores, test_f1_scores, train_recalls, test_recalls, train_precisions, test_precisions


test_dataset = PneumoniaMNIST(split='test', download=True, transform=test_transform, as_rgb=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

results = train_model(simclr_model, linear_probe, train_loader, test_loader, device, n_epochs)

train_losses, test_losses, train_f1_scores, test_f1_scores, train_recalls, test_recalls, train_precisions, test_precisions = results

# --- Визуализация ---
plot_results(
    train_losses, 
    test_losses, 
    train_f1_scores, 
    test_f1_scores, 
    train_recalls, 
    test_recalls,
    "./task3_tr4.png"
)
```
